# Q3. Sales of products in four different regions is tabulated for males and females. Find if male-female buyer rations are similar across regions.
 

```python
sales=pd.read_csv("BuyerRatio.csv")
sales
```

```python
#Chi square test
#The Null and Alternative Hypothesis

#variables are independent, H0 (proportions are equal)

#variables are dependent, Ha
```

```python
alpha=0.05
Male = [50,142,131,70]
Female=[435,1523,1356,750]
Sales=[Male,Female]
print(Sales)
```

```python
chiStats = sp.stats.chi2_contingency(Sales)
print('Test t=%f p-value=%f' % (chiStats[0], chiStats[1]))
print('Interpret by p-Value')
if chiStats[1] < 0.05:
  print('we reject null hypothesis')
else:
  print('we accept null hypothesis')
```

```python
#variables are independent, there is no relationship b/w variables.
#male-female buyer rations are  similar across regions. 
```
